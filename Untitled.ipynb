```python
import os
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter

# --- Step 1 & 2: Load PDF ---
# Upload your PDF to Colab files on the left and replace 'course_notes.pdf'
# If you don't have one yet, create a dummy one or skip to text loading.
file_path = "course_notes.pdf" 

if os.path.exists(file_path):
    loader = PyPDFLoader(file_path)
    documents = loader.load()
else:
    # Fallback if no PDF is uploaded (for testing)
    from langchain.docstore.document import Document
    text = """
    The powerhouse of the cell is the mitochondria. 
    Newton's second law is F=ma. 
    Photosynthesis takes place in the chloroplasts.
    """
    documents = [Document(page_content=text)]

# --- Step 3: Chunking ---
# We split text into 300-character chunks with some overlap to maintain context.
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=50
)
chunks = text_splitter.split_documents(documents)

print(f"Number of chunks created: {len(chunks)}")
print(f"Sample chunk: {chunks[0].page_content}")
```
